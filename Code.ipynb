```python
#Station 43279 Madras

import matplotlib.pyplot as plt
import pandas as pd


Data = pd.read_csv (r'C:\Users\Test\Desktop\Study\Sem 6\ECS\Station 43279 Madras 18th March.csv')

#print(Data)

cols = Data.columns

figure, ax1 = plt.subplots()
ax1.plot(Data[cols[0]],Data[cols[1]], color ='r', label = "Pressure (hPa)")
ax1.plot(Data[cols[2]],Data[cols[1]], color ='g', label = "Temperature (C)")
ax1.plot(Data[cols[4]],Data[cols[1]], color ='b', label = "Humidity (%)")
plt.ylabel('Height (m)')
plt.xlabel('Values')
plt.title('Height vs Pressure, Temperature, Humidity Plot for Station 43279 Madras')
plt.legend()
plt.show
plt.savefig('Station 43279 Madras', bbox_inches='tight')

figuret, ax1 = plt.subplots()
ax1.plot(Data[cols[2]],Data[cols[1]], color ='g', label = "Temperature (C)")
plt.ylabel('Height (m)')
plt.xlabel('Temperature (C)')
plt.title('Height vs Temperature Plot for Station 43279 Madras')
plt.legend()
plt.show
plt.savefig('Temperature plot Station 43279 Madras', bbox_inches='tight')

figurep, ax1 = plt.subplots()
ax1.plot(Data[cols[0]],Data[cols[1]], color ='r', label = "Pressure (hPa)")
plt.ylabel('Height (m)')
plt.xlabel('Pressure (hPa)')
plt.title('Height vs Pressure Plot for Station 43279 Madras')
plt.legend()
plt.show
plt.savefig('Pressure plot for Station 43279 Madras', bbox_inches='tight')

figureh, ax1 = plt.subplots()
ax1.plot(Data[cols[4]],Data[cols[1]], color ='b', label = "Humidity (%)")
plt.ylabel('Height (m)')
plt.xlabel('Humidity (%)')
plt.title('Height vs Humidity Plot for Station 43279 Madras')
plt.legend()
plt.show
plt.savefig('Humidity plot for Station 43279 Madras', bbox_inches='tight')
```
